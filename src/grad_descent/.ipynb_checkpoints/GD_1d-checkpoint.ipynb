```python
from __future__ import division, print_function, unicode_literals
import math
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def grad(x):
    return 2*x+ 5*np.cos(x)

def cost(x):
    return x**2 + 5*np.sin(x)

def myGD1(x0, eta):
    x = [x0]
    for it in range(100):
        x_new = x[-1] - eta*grad(x[-1])
        if abs(grad(x_new)) < 1e-3:
            break
        x.append(x_new)
    return (x, it)

(x1, it1) = myGD1(-5, .1)
(x2, it2) = myGD1(5, .1)
print('Solution x1 = %f, cost = %f, obtained after %d iterations'%(x1[-1], cost(x1[-1]), it1))
print('Solution x2 = %f, cost = %f, obtained after %d iterations'%(x2[-1], cost(x2[-1]), it2))
```

```python

from matplotlib.backends.backend_pdf import PdfPages



def draw_gd1d(x1, ids, filename):
    x0 = np.linspace(-5.5, 5.5, 1000)
    y0 = cost(x0)
    nrows = 2
    ncols = 4
    width = 16
    height = 8


    plt.close('all')
    fig,axs=plt.subplots(nrows,ncols,figsize=(width,height)) 
    with PdfPages(filename) as pdf: 
        for i, k in enumerate(ids):
            r = i//ncols 
            c = i%ncols 
            axs[r, c].plot(x0, y0, 'b')
            x = x1[ids[i]]
            y = cost(x)
            str0 = 'iter {}/{}, grad = {:.3f}'.format(ids[i], len(x1)-1, grad(x))
            axs[r, c].set_xlabel(str0, fontsize = 13)
            axs[r, c].plot(x, y, 'ro', markersize=7, markeredgecolor = 'k')
            axs[r, c].plot()

        pdf.savefig(bbox_inches='tight')
        
        plt.show()
        
# good learning rate 
filename = 'gd1d_0.pdf'
(x1, it0) = myGD1(-5, .1)
ids = [0, 1, 2, 3, 4, 5, 7, 11]
draw_gd1d(x1, ids, filename)
```

```python
# small learning rate 
filename = 'gd1d_1.pdf'
(x1, it0) = myGD1(-5, .01)
ids = [0, 10, 20, 30, 50, 70,  90, 100]
draw_gd1d(x1, ids, filename)
```

```python
# small learning rate 
filename = 'gd1d_2.pdf'
(x1, it0) = myGD1(-5, .5)
ids = [0, 10, 20, 30, 50, 70,  90, 100]
draw_gd1d(x1, ids, filename)
```

```python
# good learning rate 
filename = 'gd1d_3.pdf'
(x1, it0) = myGD1(5, .1)

ids = [0, 3, 6, 10, 15, 20, 25, 29]
draw_gd1d(x1, ids, filename)
```

```python
from sklearn.linear_model import LinearRegression 
X = np.random.rand(1000, 1)
y = 4 + 3 * X + .5*np.random.randn(1000, 1) # noise added

model = LinearRegression()
model.fit(X, y)

w, b = model.coef_[0][0], model.intercept_[0]
sol_sklearn = np.array([b, w])
print('Solution found by sklearn:', sol_sklearn)

# Draw the fitting line 
x0 = np.linspace(0, 1, 2, endpoint=True)
y0 = w*x0 + b
filename = 'LR_data.pdf'
with PdfPages(filename) as pdf: 
    plt.plot(X.T, y.T, 'b.')     # data 
    plt.plot(x0, y0, 'y', linewidth = 4)   # the fitting line
    plt.axis([0, 1, 0, 10])
    pdf.savefig(bbox_inches = 'tight')
    plt.show()
```

```python
# Building Xbar 
one = np.ones((X.shape[0],1))
Xbar = np.concatenate((one, X), axis = 1)

def grad(w):
    N = Xbar.shape[0]
    return 1/N * Xbar.T.dot(Xbar.dot(w) - y)

def cost(w):
    N = Xbar.shape[0]
    return .5/N*np.linalg.norm(y - Xbar.dot(w), 2)**2
```

```python
def check_grad(fn, gr, X):
    # convert X to an 1d array, later we'll need only one for loop
    X_flat    = X.reshape(-1)
    shape_X   = X.shape                # original shape of X 
    num_grad  = np.zeros_like(X)       # numerical grad, shape = shape of X 
    grad_flat = np.zeros_like(X_flat)  # 1d version of grad
    eps       = 1e-6            # a small number, 1e-10 -> 1e-6 is often good
    numElems  = X_flat.shape[0] # number of elements in X 
    # calculate numerical gradient 
    for i in range(numElems):          # iterate over all elements of X 
        Xp_flat      = X_flat.copy()
        Xn_flat      = X_flat.copy()
        Xp_flat[i]  += eps
        Xn_flat[i]  -= eps
        Xp           = Xp_flat.reshape(shape_X) 
        Xn           = Xn_flat.reshape(shape_X)
        grad_flat[i] = (fn(Xp) - fn(Xn))/(2*eps)

    num_grad = grad_flat.reshape(shape_X) 
    
    diff = np.linalg.norm(num_grad - gr(X))
    print('Difference between two methods should be small:', diff)
```

```python
check_grad(cost, grad, np.random.randn(2, 1))
```

```python
def myGD(w_init, grad, eta):
    w = [w_init]
    for it in range(100):
        w_new = w[-1] - eta*grad(w[-1])
        if np.linalg.norm(grad(w_new))/len(w_new) < 1e-3:
            break 
        w.append(w_new)
    return (w, it)

w_init = np.array([[2], [1]])
(w1, it1) = myGD(w_init, grad, 1)
print('Sol found by GD: w = ', w1[-1].T, ',\nafter %d iterations.' %(it1+1))
```
